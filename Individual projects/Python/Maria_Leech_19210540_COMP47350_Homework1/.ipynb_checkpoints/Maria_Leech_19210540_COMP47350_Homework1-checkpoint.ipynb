## Transformation of features

In this first new feture I have combined if a patient is 80+ years and not in hospital their chances of death as a result of Covid-19 are higher than any other group.

```python
#adding a new column death_risk_prediction_age based on hospital admission. if hospitalised and over 70 there is a greater chance of death. 
newdf['death_risk_prediction_age'] = np.where((newdf['age_group']== '80+ Years') | (newdf['age_group']== '70 - 79 Years') & (newdf['hosp_yn']=='Yes'), "High", "Low")
newdf.head()
```

```python
#adding a new column that predict whether you've a high or low chance of death based on age and race.
newdf['death_risk_prediction_race'] = np.where(((newdf['race_ethnicity_combined']== 'White, Non-Hispanic') & (newdf['age_group']=='80+ Years') | (newdf['age_group']=='70 - 79 Years')) | ((newdf['race_ethnicity_combined']== 'Black, Non-Hispanic') & (newdf['age_group']=='80+ Years') | (newdf['age_group']=='70 - 79 Years')) , "High", "Low")
newdf.head()
```

```python
#count based on death and gender and age group
gender_count = newdf.groupby(['death_yn','sex','age_group']).size()
gender_count
```

Based on previous death rate analysis men are more likely to die as a result of covid-19. 
31.9% chance of death female in 80+ group
40.9% chance of death male in the 80+ group


```python
#adding a new column that predict whether you've a high or low chance of death based on age and sex.
newdf['death_risk_prediction_gender_&_age_80+'] = np.where(((newdf['sex']=='Male') & (newdf['age_group']== '80+ Years') | (newdf['age_group']== '70 - 79 Years')) | ((newdf['sex']=='Female') & (newdf['age_group']== '80+ Years') | (newdf['age_group']== '70 - 79 Years') ), "High", "Low")
newdf.head()
```

```python
#New target column that based on previous three colunms says that if you got a high in two or more of the other columns you've a higher chance of death as a result of Covid-19
newdf['death_risk_prediction'] = np.where((newdf['death_risk_prediction_age']=='High') & (newdf['death_risk_prediction_race']== 'High') & (newdf['death_risk_prediction_gender_&_age_80+']== 'High') | ((newdf['death_risk_prediction_age']=='Low') & (newdf['death_risk_prediction_race']== 'High') & (newdf['death_risk_prediction_gender_&_age_80+']== 'High')) |((newdf['death_risk_prediction_age']=='High') & (newdf['death_risk_prediction_race']== 'Low') & (newdf['death_risk_prediction_gender_&_age_80+']== 'High')) | ((newdf['death_risk_prediction_age']=='High') & (newdf['death_risk_prediction_race']== 'High') & (newdf['death_risk_prediction_gender_&_age_80+']== 'Low')), "Very-High", "Low")
newdf.head()
```

### Conclusion

I added three new features to the dataset death_risk_prediction_age, death_risk_prediction_race, death_risk_prediction_gender_&_age_80+. This allows use to categorise the groups based on there age gender race and a combination of these features to predict the likelihood of an adverse outcome for the patients in these high risk groups. 
I used these three new columns to create a fourth column which is a new target feature and helps to predict if a patient is more at risk of death based on age gender and race. If they score a high in two or more they have a higher chance of dying from Covid-19 than patients in the other groups in this dataset.
I also want to mention that this is just a subset of data from a database containing 18 million data rows and so this may not be an indicative report that could be applied to the full dataset. 
Going forward I would also suggest that we review the race ethnicity values and prepare a data sample that has a more balanced approach as White Non-Hispanic group had the largest representation triple that of other ethnic groups. We could under sample the White Non-Hispanic groups or over sample the other groups while still maintaing a good balace across genders and age groups.
From the analysis of this data I conclude that you have a higher chance of death if you are over 70 years of age and are of Black or White ethnicity/race, while race and gender do contribute the largest factor is age and anybody in the 70+ age group needs to continue to be cautious and avoid the risk of contracting Covid-19.
